# Analyzing multiple columns

```python
sns.violinplot(x='Sex',y='HeartDisease',data=df)
```

Conclusion:-More males are having heart disease as compared to female.

```python
sns.violinplot(y=df['Age'],x=df['HeartDisease'])
```

Conclusion:Average age of not having heart disease is around 50 & of having heart disease is around  58-60.

```python
sns.violinplot(x=df['ChestPainType'],y=df['HeartDisease'])
```

Conclusion:-For,ASY Chest Pain there is high probability of having heart disease.

```python
sns.violinplot(x=df['FastingBS'],y=df['HeartDisease'])
```

Conclusion:-There is high possibility of having Heart Disease with high FastingBS.
